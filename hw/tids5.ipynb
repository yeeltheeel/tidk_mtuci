```python
import numpy as np
from scipy.stats import binom, entropy
from math import log2
```

```python
# Задача 5.1 
# Один из символов x \in A=\{a_1 ... a_L\} дискретного источника информации (ДИИ) появляется на его выходе с вероятностью p=P\{x=a_k\}:
# 1) p=0.1; 
# 2) p=0.27; 
# 3) p=0.05; 
# 4) p=0.46.
# Найти его собственную информацию I(x) в битах. 

p = [0.1, 0.27, 0.05, 0.46]

for pi in p:
   print("p = {}: I = {:.4f}  бит".format(pi, -log2(pi)))
```

```python
#  Задача 5.2
# Дискретный двоичный источник без памяти (ДИИ-БП) выдает случайные символы X={x_i}, i=1,2,3,..., x_i \in A={0,1} с вероятностями p и q. 
# 1) Найти энтропию каждого символа источника, если 
#     а) p=0.1; 
#     б) p=0.3; 
#     в) p=0.5; 
#     г) p=0.7; 
#     д) p=0.9; 
# 2) Найти энтропию ДИС-БП из 7 символов этого источника для тех же вероятностей. 
# 3) Найти максимальную энтропию сообщения, состоящего, соответственно, из 
#     а) 5; 
#     б) 14; 
#     в) 23; 
#     г) 50 
# символов. 

p = [0.1, 0.3, 0.5, 0.7, 0.9]
k = 7
n = [5, 14, 23, 50]

print("1)")
h = []
for pi in p:
    qi = 1 - pi
    hx = -pi * log2(pi) - qi * log2(qi)
    h.append(hx)
    print("p = {}: H = {:.4f} бит".format(pi, hx))

print("2)")
for x in range(len(p)):
    print("p = {}: H = {:.4f} бит".format(p[x], k * h[x]))

print("3)")
for ni in n:
    pi = 0.5
    print("n = {}: Hmax = {} бит".format(ni, ni))
```

```python
#  Задача 5.3
# ДИИ-БП имеет объем алфавита L и формирует ДИС в виде бесконечной последовательности отсчетов. Чему равна максимально возможная энтропия: 
# а) каждого символа такого источника; 
# б) сообщения, составленного из нескольких символов, если число символов в сообщении равно 
#     а) 3; 
#     б) 10; 
#     в) 27.

n = [3, 10, 27]
L = 1e10

print("а)")
hx = log2(L)
print("Hmax = {:.4f}".format(hx))

print("б)")
for ni in n:
    print("n = {}: Hmax = {:.4f} бит".format(ni, ni * hx))
```

```python
#  Задача 5.4
# Даны три дискретных источника без памяти:
# 1) у алфавита L=2, символы алфавита равновероятны; 
# 2) у алфавита L=3, символы источника равновероятны;
# 3) у алфавита L=2, символы появляются с вероятностями p=0.3, q=0.7. 
# Вычислить энтропию всех этих источников и определить, какой из них обладает большей энтропией.

L = [2,   3,   2]
p = [0.5, 0.5, 0.3]
q = [0.5, 0.5, 0.7]

h_max = 0
for i in range(len(L)):
    print("{}) L = {} бит, ".format(i+1, L[i]), end="")
    if (p[i] == q[i]):
        hx = log2(L[i])
        h_max = max(h_max, hx)
        print("H = {:.4f} бит".format(hx))
    else:
        hx = -p[i] * log2(p[i]) - q[i] * log2(q[i])
        h_max = max(h_max, hx)
        print("H = {:.4f} бит".format(hx))
print("Hmax = {:.4f} бит".format(h_max))
```

```python
#  Задача 5.5 
# Дискретный источник информации выдает случайные символы X={x}, x \in A={0,1, ..., n} 
# с биномиальным законом распределения p(k)=P_n({x=k})=C_n^k*p^k*(1-p)^(n-k), \forall k \in A, C_n^k=\frac{n!}{k!(n-k)!}.
# Определить энтропию H[X] этого источника: 
# а) в общем случае; 
# б) при p=0.5 и n=5; 
# в) p=0.1 и n=4.

print("{}) H[X] = {:.4f} бит".format('a', 0))

p = [0.5, 0.1]
n = [5, 4]
let = 'б'
for i in range(len(p)):
    probs5b = binom.pmf(np.arange(6), n[i], p[i])
    h5b = entropy(probs5b, base=2)
    print("{}) H[X] = {:.4f} бит".format(let, h5b))
    let = chr(ord(let) + 1)
```

```python
#  Задача 5.6 
# Дискретное распределение вероятностей (ДРВ) случайного ДИС X={x}, x \in A={a_1, a_2, a_3, a_4} имеет вид p(a_k)=[0.1, 0.1, 0.1, 0.7]. 
# Определить какое число M элементов алфавита B={b_1, ..., b_M} должно быть у ДИС Y={y}, Y \in B с равномерным ДРВ случайной величины y, 
# чтобы энтропия ДИС Y была наиболее близка к энтропии ДИС X.

p = np.array([0.1, 0.1, 0.1, 0.7])

hx = -np.sum([pi * log2(pi) for pi in p if pi > 0])
print("Hx = {:.4f} бит".format(hx))
# Hy = log2(M)
m = round(pow(2, hx))
print("M = {}".format(m))
```

```python
#  Задача 5.7
# Для двух случайных ДИС X={x}, x \in A=\{a_1, a_2, a_3\} и Y=\{y\}, y \in B=\{b_1, b_2\} известно совместное взаимное ДВР: 
# p(a_k, b_r) = 
# \\begin{cases}
# 	p(a_1, b_1)=0.1; p(a_1, b_2)=0.25; p(a_2, b_1)=0.2; & \text{} \\
# 	p(a_2, b_2)=0;   p(a_3, b_1)=0.3;  p(a_3, b_2)=0.15; & \text{} 
# \\end{cases}.
p = np.array([
    [0.1, 0.25],
    [0.2, 0],
    [0.3, 0.15]
])
# Определить: 
# а) собственные энтропии H[X],H[Y]; 
# б) совместную взаимную энтропию H[X,Y]; 
# в) условные взаимные энтропии H[X|Y], H[Y|X]; 
# г) взаимную энтропию H[X \wedge Y].

pF = p.sum()
pX = p.sum(axis=1) # по столбцам
pY = p.sum(axis=0) # по строкам
# print("check:", pF, pX.sum(), pY.sum())

Hx = -np.sum([px * log2(px) for px in pX if px > 0])
Hy = -np.sum([py * log2(py) for py in pY if py > 0])
print("а) H[X] = {:.4f} бит, H[Y] = {:.4f} бит".format(Hx, Hy))

Hxy = -np.sum([pij * log2(pij) for pij in p.flatten() if pij > 0])
print("б) H[X,Y] = {:.4f} бит".format(Hxy))

Hxy1 = Hxy - Hy
Hyx1 = Hxy - Hx
print("в) H[X|Y] = {:.4f} бит, H[Y|X] = {:.4f} бит".format(Hxy1, Hyx1))

Hxy2 = Hx + Hy - Hxy
print("г) H[X∧Y] = {:.4f} бит".format(Hxy2))
```

```python
# Задача 5.8
# ДИИ формирует двухмерное случайное ДИС X=\{x_1, x_2\}, x_i \in A=\{a_1, a_2, a_3, a_4\} из двух СВ x=\{x_1, x_2\}, с алфавитом A из 4-х символов. 
# Известны одномерные безусловные p_1(a_k1)=P(\{x=a_k1\}), \forall a_k \in A, и условные p_1(a_k2|a_k1)=P(\{x_2=a_k2|x_1=a_k1\}), 
# \forall a_k1,a_k2 \in A, дискретные вероятностные распределения (ДВР) этих величин этих СВ, заданные ниже двумя таблицами:
p1 = np.array([0.5, 0.25, 0.125, 0.125])
a = np.array([
    [5/8, 3/8, 0,   0],
    [1/8, 1/2, 3/8, 0],
    [0,   0,   0,   1],
    [1/2, 1/4, 1/4, 0]
])
L = 4
# 1) Найти собственные энтропии ДИС: H[x1, x2], H[x1|x2], H[x2|x1], H[x1 \wedge x2]; 
# 2) Найти производительность, насыщенность и избыточность источника: J_p[x], J_s[x], J_r[x] (T=0.1с)

p = np.zeros((len(p1), len(a[0])))
for i in range(len(p1)):
    for j in range(len(a[0])):
        p[i, j] = p1[i] * a[i][j]

# check
# for k in range(len(p)):
#     for r in range(len(p[0])):
#         print(p[k, r], end=' ')
#     print()
# pa = a.sum(axis=0)
# pb = a.sum(axis=1)
# print("pa, pb:", pa, pb)
# print()

hx1 = -np.sum([pi * log2(pi) for pi in p1 if pi > 0])
hx2 = -np.sum([pi * log2(pi) for pi in p.sum(axis=0) if pi > 0])
hx1x2 = -np.sum([pi for pi in p.flatten() if pi > 0] * np.log2([pi for pi in p.flatten() if pi > 0]))
# hx1x2_lp = 0
# for k in range(len(p)):
#     for r in range(len(p[0])):
#         hx1x2_lp -= p[k, r] * log2(p[k, r]) if p[k, r] > 0 else 0
# print("hx1x2:", hx1x2, hx1x2_lp)

hx_arr = [hx1x2, hx1x2 - hx2, hx1x2 - hx1, hx1 + hx2 - hx1x2]
print("H[x1,x2] = {:.4f} бит".format(hx_arr[0]))
print("H[x1|x2] = {:.4f} бит".format(hx_arr[1]))
print("H[x2|x1] = {:.4f} бит".format(hx_arr[2]))
print("H[x1∧x2] = {:.4f} бит".format(hx_arr[3]))

T = 0.1
hmax = log2(L)
j = [hx1x2/T, hx1x2/hmax, 1 - hx1x2/hmax]
print("Jp[X] = {:.4f} бит/с".format(j[0]))
print("Js[X] = {:.4f}".format(j[1]))
print("Jr[X] = {:.4f}".format(j[2]))
```

```python
# Задача 5.9
# Дискретный стационарный марковский источник формирует последовательность случайных величин {x_i}, x_i \in A={a1,a2,a3,a4}, с алфавитом A из 4-х 
# символов. Известны одномерные безусловные p1(a_ki)и условные переходные p(a_ki|a_{k_{i-1}}) дискрет. вероят. распределения (ДВР) этого марковского 
# процесса, приведенные ниже в соответствующих таблицах:
p1 = [0.6, 0.2, 0.15, 0.05]
a = [
    [0.8,  0.1,  0.1, 0   ],
    [0.2,  0.6,  0.2, 0   ],
    [0,    0.25, 0.7, 0.05],
    [0.05, 0.05, 0.4, 0.5 ]
]
L = 4
# Вычислить энтропию H[x] из n элементов такого марковского сообщения и соответствующую избыточность источника Jr[x] при T=0.1 с. 
# Вычислить эти же показатели H[x], Jr[x] для аналогичного сообщения без памяти c теми же безусловными ДВР и сравнить полученные результаты с  
# предыдущими.Расчеты выполнить для двух вариантов: 
# 1) n=2;
# 2) n=3.

T = 0.1

hx0 = 0
for i in range(len(p1)):
    hx0 -= p1[i] * np.sum([pk * log2(pk) for pk in a[i] if pk > 0])
hx1 = -np.sum([pi * log2(pi)  for pi in p1])

for n in range(2, 4):
    hmax = n * log2(L)
    print("{}) n = {}:".format(n-1, n))
    hx = hx1 - (n - 1) * hx0
    jr = 1 - hx/hmax
    print("H[x] = {:.4f} бит, Jr[x] = {:.4f}".format(hx, jr))
    hxm = n * hx1
    jrm = 1 - hxm/hmax
    print("H[x] = {:.4f} бит, Jr[x] = {:.4f} (no mem)".format(hxm, jrm))
```

```python
# Задача 5.10
# Решить задачу 5.9. для случая, когда стационарный дискретный марковский источник формирует бинарную последовательность {x_i}, x_i \in A={a1,a2} с 
# алфавитом A из 2-х символов, а соответствующие одномерные безусловные и условные ДВР задаются таблицами:
p = [0.3, 0.7]
a = [
    [0.4, 0.6],
    [0.1, 0.9]
]
L = 2

# предположим, что n остаются неизменными
hx0 = 0
for i in range(len(p)):
    hx0 -= p1[i] * np.sum([pk * log2(pk) for pk in a[i] if pk > 0])
hx1 = -np.sum([pi * log2(pi)  for pi in p])
for n in range(2, 4):
    hmax = n * log2(L)
    print("{}) n = {}:".format(n-1, n))
    hx = hx1 - (n - 1) * hx0
    jr = 1 - hx/hmax
    print("H[x] = {:.4f} бит, Jr[x] = {:.4f}".format(hx, jr))
    hxm = n * hx1
    jrm = 1 - hxm/hmax
    print("H[x] = {:.4f} бит, Jr[x] = {:.4f} (no mem)".format(hxm, jrm))
```

```python
# Задача 5.11
# Дискретный источник информации без памяти с объемом алфавита L=5 обладает энтропией H(X)=3.5. 
# Найти производительность, насыщенность и избыточность данного источника, если время формирования одного символа источником занимает Tэ=0.1с.

L = 5
Hx = 3.5
T = 0.1

j = [Hx/T, log2(L)/T, 1 - Hx/log2(L)] # Jr=1-Jp/Js
print("Jp[X] = {:.4f} бит/с".format(j[0]))
print("Js[X] = {:.4f}".format(j[1]))
print("Jr[X] = {:.4f}".format(j[2]))
```

```python
# Задача 5.12
# Стационарный дискретный случайный процесс задан распределением:
x = [-2, -1, 0, 1, 2, 3, 4, 5]
p = [0.05, 0.05, 0.08, 0.1, 0.12, 0.15, 0.2, 0.25]
# Здесь x_i - возможные значения процесса, p_i - их вероятности. 
# Найти математическое ожидание и дисперсию процесса.

mx = 0
for i in range(len(x)):
    mx += x[i] * p[i]
dx = 0
for i in range(len(x)):
    dx += (x[i] - mx) * (x[i] - mx) * p[i]
print("MX = {:.4f}\nDX = {:.4f}".format(mx, dx))
```

```python
# Задача 5.13
# Определите апостериорные вероятности передачи символов p(x_i|y_j) в двоичном дискретном канале связи, если канал задан матрицей переходных 
# вероятностей:
# P = 
# \begin{matrix}
#     p(y_1|x_1) & p(y_1|x_2) \\
#     p(y_2|x_1) & p(y_2|x_2)
# \end{matrix}
p = np.array([
    [0.9, 0.1], 
    [0.1, 0.9]
])
# p(x1)=0.8, p(x2)=0.2.
px = [0.8, 0.2]
# Ответ дать в виде матрицы 
# \begin{matrix}
#     p(x_1|y_1) & p(x_1|y_2) \\
#     p(x_2|y_1) & p(x_2|y_2)
# \end{matrix}

# P(x|y) = P(y|x)*P(x)/P(y)
py = np.zeros(len(p))
for i in range(len(p)):
    for j in range(len(p[0])):
        py[i] += p[i, j] * px[j]

p1 = np.zeros((len(px), len(py)))
for i in range(len(p)):
    for j in range(len(p[0])):
        p1[i, j] = p[i, j] * px[i] / py[j]

for i in range(len(p1[0])):
    for j in range(len(p1)):
        print("{:.4f}".format(p1[i, j]), end=" ")
    print()
```
